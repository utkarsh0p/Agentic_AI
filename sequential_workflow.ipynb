```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class MultiplyState(TypedDict):
    number_one:int
    number_two:int
    result:int
```

```python
def multiply(state: MultiplyState):
    result = state["number_one"]*state["number_two"]
    return {'result':result}
```

```python
graph = StateGraph(MultiplyState)
```

```python
graph.add_node("multiply", multiply)
```

```python
graph.add_edge(START, 'multiply')
graph.add_edge('multiply', END)
```

```python
workflow = graph.compile()
```

```python
workflow.invoke({'number_one':3, "number_two":4})
```
